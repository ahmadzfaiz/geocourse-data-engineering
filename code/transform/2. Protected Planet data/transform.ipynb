## Unified process for all countries

```python
countries = ["BRN", "SGP", "TLS"]
```

```python
for country in countries:
  print("Processing", country)
  with zipfile.ZipFile(f"../../extract/output/PA_{country}.zip", 'r') as zip_ref:
    # Extract only the spatial file
    for file in zip_ref.namelist():
      if f"WDPA_WDOECM_Feb2025_Public_{country}_shp" in file:
        zip_ref.extract(file, "temp")

    list_gdf = []
    for number in list(range(0, 3)):
      gdf = gpd.read_file(f"temp/WDPA_WDOECM_Feb2025_Public_{country}_shp_{number}.zip")
      gdf["source"] = "Protected Planet"
      list_gdf.append(gdf)

    merged_gdf = pd.concat(list_gdf, ignore_index=True)
    merged_gdf.to_file("../output.gpkg", layer=f"{country.lower()}_protected_areas", driver="GPKG")
```
