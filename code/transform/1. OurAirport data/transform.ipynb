```python
import pandas as pd
import geopandas as gpd
from shapely import Point
```

```python
df = pd.read_csv("../../extract/output/airports_indonesia.csv")

df.head()
```

```python
df.info()
```

```python
df.head(1).apply(
  lambda row: print(row), 
  axis=1
)
```

```python
df.head(1).apply(
  lambda row: print(
    row["longitude_deg"], 
    row["latitude_deg"],
    row["elevation_ft"]
  ), 
  axis=1
)
```

```python
df["geometry"] =  df.apply(
  lambda row: Point(
    row["longitude_deg"], 
    row["latitude_deg"],
    row["elevation_ft"]
  ), 
  axis=1
)
```

```python
df.head()
```

```python
df.info()
```

```python
gdf = gpd.GeoDataFrame(df, geometry='geometry')

gdf.head()
```

```python
gdf.info()
```

```python
gdf.crs
```

```python
gdf.set_crs(epsg=4326, inplace=True)
```

```python
gdf.crs
```

```python
gdf["source"] = "OurAirports"
```

```python
gdf.to_file("../output.gpkg", layer="airports_indonesia", driver="GPKG")
```
